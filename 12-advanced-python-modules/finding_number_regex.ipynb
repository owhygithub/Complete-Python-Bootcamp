```python
def get_text_from_file(directory, file):
    path = directory + "/" + file
    print(f"This is the path - {path}")
    try:
        with open(path, 'r', encoding='utf-8') as f:
            # print(f.read())
            return f.read()
    except UnicodeDecodeError:
        with open(path, 'r', encoding='latin-1') as f:
            # print(f.read())
            return f.read()
```

```python
counter = 0
for folder, subfolders, files in os.walk(f"{os.getcwd()}/extracted_content"):
    print(f"\tLooking at folder: {folder.split("/")[-1]}.")
    for subfolder in subfolders:
        print(f"Exploring subfolder: {subfolder}.")

    counter = 0
    for file in files:
        print(f"\t\tSearching for pattern in file: {file}.")
        text = get_text_from_file(folder, file)
        print(text)
        bingo = re.search(pattern, text)
        print(bingo)
        if bingo is None:
            counter += 1
            continue
        else:
            print(f"Found phone number {bingo} in folder {folder} and file {file}.")
            break
```
