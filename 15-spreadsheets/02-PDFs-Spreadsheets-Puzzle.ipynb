## Task One: Grab the Google Drive Link from .csv File

```python
import pypdf
```

```python
file = open('exercise-files/Find_the_phone_Number.pdf', 'rb')
```

```python
pdf_file = pypdf.PdfReader(file)
```

```python
number_of_pages = len(pdf_file.pages)
number_of_pages
```

```python
import re
```

```python
pattern = r'\d{3}' # pattern of 3 digits
```

```python
all_text = ''

for x in range(number_of_pages):
    text = pdf_file.pages[x].extract_text()
    # print(type(text))
    all_text += '' + text
```

```python
all_text
```

```python
match = re.findall(pattern, all_text)
```

```python
match
```

```python
# general format 505 503 445
known_pattern = r'\d\d\d.\d\d\d.\d\d\d\d'
exact_match = re.search(known_pattern, all_text)
```

```python
print(exact_match.group())
```
