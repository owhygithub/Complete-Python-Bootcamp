```python
nested_list = []
for _ in range(int(input())):
    name = input()
    score = float(input())
    couple = [name, score]
    nested_list.append(couple)
        
    scores = [x[1] for x in nested_list]
    unique_scores = sorted(set(scores))
    second_lowest = unique_scores[1] # got the second lowest score
    students = sorted([x[0] for x in nested_list if x[1] == second_lowest])

    # NOTE new printing method without using for loop
    print(*students, sep="\n")
    
    # for x in students:
    #     print(x)
```
